```python
# data from origin csv 
ddd = dataset
```

```python
ddd
```

```python
# delete nan in row (row means horizantal direction)
ddd = ddd.dropna()
```

```python
#Check array dimension
ddd.shape
```

```python
# transform DataFrame data to Numpy array
ddd = ddd.values
```

```python
# Loop all 2-d array element and transform all the elements to str type data
ii = 0
for x in range(0, 62):
    for y in range(0, 13):
        ddd[x, y] = str(ddd[x, y])
        ii +=1
```

```python
ii
```

```python
62*13
```

```python
# Loop all 2-d array element and filter all the elements that are special pattern or symbol.
ii = 0
for x in range(0, 62):
    for y in range(0, 13):
        ddd[x, y] = re.sub('[^A-Za-z0-9\u4e00-\u9fa5]', '',  ddd[x, y])
        ii +=1
```

```python
ddd
```

```python
# Transform numpy data back to DataFrame data
ddd = pd.DataFrame(ddd)
```
